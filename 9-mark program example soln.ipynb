```python
# 11
pd.crosstab(df['Sex'],df['Survived'])
```

```python
# 12
age_grp=[]
for i in df['Age']:
    if i<=18:
        age_grp.append('child')
    elif i<=30:
        age_grp.append('Young')
    elif i<=58:
        age_grp.append('Adult')
    else:
        age_grp.append('Senior Citizen')
        
df['age_group']=age_grp
print(df)
```

```python
# 14
from pandas.plotting import parallel_coordinates
pll=parallel_coordinates(df,'Survived',cols=['Pclass','Fare','Age'],color=('red','blue','green','yellow','orange'))
plt.show()
```

```python
# 15
sns.set(font_scale=1.5)
sns.regplot(x='Age',y='Fare',data=df,color='green',marker='+',scatter_kws={'s':200})
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
# 16
cols = df[['Age', 'Fare','Survived']]
pd.plotting.scatter_matrix(cols,figsize=[10,10],marker='o',alpha=0.7,color='red')
plt.show()
```

```python
# 17
males = df[df['Sex']=='male'].shape[0]

males_survive = df[(df['Sex']=='male') & (df['Survived']==1)].shape[0]

print("Male survival rate",males/males_survive)

females = df[df['Sex']=='female'].shape[0]

female_survive = df[(df['Sex']=='female') & (df['Survived']==1)].shape[0]

print("Female survival rate",females/female_survive)
```
