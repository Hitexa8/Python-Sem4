```python
# Multiple linear regression
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Downloads/car data.csv")

df.info()

df['Age'] = 2024-df['Year']

x = df[['Age','Kms_Driven','Fuel_Type','Seller_Type','Transmission']]
y = df['Selling_Price']
x = pd.get_dummies(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)

print(x_train,x_train.shape)
print(x_test,x_test.shape)
print(y_train,y_train.shape)
print(y_test,y_test.shape)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
diff=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
print(diff)

from sklearn import metrics
print("MAE",metrics.mean_absolute_error(y_test,y_pred))
print("MSE",metrics.mean_squared_error(y_test,y_pred))
print('R2Score',metrics.r2_score(y_test,y_pred))

lr.coef_

lr.intercept_
```

```python
# Multiple linear regression
df1 = pd.read_csv("Downloads/Advertising.csv")

x = df1[['TV','Radio','Newspaper']]
y = df1['Sales']

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
diff=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
print(diff.to_string())

from sklearn import metrics

print("MAE",metrics.mean_absolute_error(y_test,y_pred))
print("MSE",metrics.mean_squared_error(y_test,y_pred))
print('R2Score',metrics.r2_score(y_test,y_pred))

lr.intercept_

lr.coef_
```

```python
# polynomial regression
df2 = pd.read_csv('Downloads/data_poly.csv')

x = df2.iloc[:,1:2]
y = df2.iloc[:,2]

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3)

model_poly = poly.fit_transform(x)

lr = LinearRegression()
lr.fit(model_poly,y)

plt.scatter(x,y)

plt.plot(x,lr.predict(model_poly))

pred = lr.predict(poly.fit_transform([[70]]))

print(pred)
```
